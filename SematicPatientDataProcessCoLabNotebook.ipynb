## Read All CSV Files

```python
patient_mapping=pd.read_csv("patient_mapping.csv")
proccessed_data=pd.read_csv("processed_data.csv")
notes=pd.read_csv("notes.csv")
admission=pd.read_csv("admission.csv")
```

## Take Notes a default Patient Data Frame

```python
patient_df=notes
```

### Get patient number and other info from registration number

```python
patient_df['dept']= [dpt[0:2] for dpt in patient_df["reg_number"].to_list()]
patient_df['patient_number']= [dpt[2:8] for dpt in patient_df["reg_number"].to_list()]
patient_df['visit_year']= [dpt[8:] for dpt in patient_df["reg_number"].to_list()]
patient_df['patient_number']=patient_df['patient_number'].astype('int64')
```

### Map Citizen number on patient number on Pateint Dataframe

```python
patient_df=pd.merge(patient_df, patient_mapping, on="patient_number")
```

###Map Patient Processed Data into Patient Data Frame based on patient number

```python
patient_df=pd.merge(patient_df, proccessed_data[['processed_by',	'patient_number',	'admission_date',	'admission_time',	'admission_type',	'final_notes',	'final_remarks']], on="patient_number")
patient_df = patient_df.rename({'admission_date': 'proccessed_admission_date', 'admission_time': 'proccessed_admission_time','admission_type':'proccessed_admission_type'}, axis='columns')
```

### Map Patient Addmission Data in to Patient Data Frame based on citizen_number

```python
patient_df=pd.merge(patient_df, admission[['citizen_number', 'admission_date', 'admision_type', 'discharge_date', 'admission_notes']], on="citizen_number")
```

```python
if "id" in  patient_df:
  del patient_df["id"]
patient_df
```

### Process and explore Possible Duplicate Data


```python
#convert to datetime object
patient_df['proccessed_admission_date'] =  pd.to_datetime(patient_df['proccessed_admission_date']).dt.date
patient_df['admission_date'] =  pd.to_datetime(patient_df['admission_date']).dt.date
```

 **if** Admission type equal processed admision type, del duplicate **column**

```python
if patient_df['proccessed_admission_type'].equals(patient_df['admision_type']):
  del patient_df['proccessed_admission_type']
```

### Check If proccessed admission date is not equal admission *date* 

```python
patient_df['proccessed_admission_date'].equals(patient_df['admission_date'])
```

```python
patient_df['proccessed_admission_date'].where(patient_df['proccessed_admission_date']!=patient_df['admission_date']).unique()
```

this mean all the rows in the processed date are equal to admission date except where nan, which mean unproccessed patients. hence add a boolean column that patient is proccesed or not 

```python
patient_df['proccessed_admission_date']==patient_df['admission_date']
```

```python
patient_df['is_patient_proccessed']=(patient_df['proccessed_admission_date']==patient_df['admission_date']).to_list()
del patient_df['proccessed_admission_date']
```
